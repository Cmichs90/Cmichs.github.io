```python
import pandas as pd

# Load Option data

df = pd.read_csv("/content/GWP1_Option_data.csv")
df.columns = ["Days", "Strike", "Price", "Type"]
df["T"] = df["Days"] / 250

## Market Inputs
S0 = 232.90
r = 0.015

## Filtering for 60 days options

df60 = df[df["Days"] == 60].copy().reset_index(drop=True)

## Call & Puts
calls60 = df60[df60["Type"] == "C"].sort_values("Strike").reset_index(drop=True)
puts60 = df60[df60["Type"] == "P"].sort_values("Strike").reset_index(drop=True)


print("60 days option data")
print(f"60 day call options: {len(calls60)}")
print(f"60 days put options: {len(puts60)}")
print(df.to_string())
```

```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import brute, fmin

def H93putvalueparity(S0, K, T, r, kappav, thetav, sigmav, rho, v0):
    call = H93callvalue(S0, K, T, r, kappav, thetav, sigmav, rho, v0)
    put = call - S0 + K * np.exp(-r * T)
    return max(0.0, float(put))

def H93charfunc(u, T, r, kappav, thetav, sigmav, rho, v0):
    c1 = kappav * thetav
    c2 = -np.sqrt((rho * sigmav * u * 1j - kappav)**2 + (sigmav**2) * (1j * u + u**2))
    c3 = (kappav - rho * sigmav * u * 1j - c2) / (kappav - rho * sigmav * u * 1j + c2)
    H1 = r * u * 1j * T + (c1 / sigmav**2) * ((kappav - rho * sigmav * u * 1j - c2) * T - 2 * np.log((1 - c3 * np.exp(-c2 * T)) / (1 - c3)))
    H2 = (kappav - rho * sigmav * u * 1j - c2) / sigmav**2 * (1 - np.exp(-c2 * T)) / (1 - c3 * np.exp(-c2 * T))
    return np.exp(H1 + H2 * v0)

def H93intfunc(u, S0, K, T, r, kappav, thetav, sigmav, rho, v0):
    charfuncvalue = H93charfunc(u - 1j * 0.5, T, r, kappav, thetav, sigmav, rho, v0)
    intfuncvalue = (1 / (u**2 + 0.25)) * np.exp(1j * u * np.log(S0 / K)) * charfuncvalue
    return intfuncvalue.real

def H93callvalue(S0, K, T, r, kappav, thetav, sigmav, rho, v0):
    intvalue, _ = quad(lambda u: H93intfunc(u, S0, K, T, r, kappav, thetav, sigmav, rho, v0), 0, np.inf, limit=250)
    callvalue = max(0.0, S0 - np.exp(-r * T) * np.sqrt(S0 * K) / np.pi * intvalue)
    return callvalue

def H93putvalueparity(S0, K, T, r, kappav, thetav, sigmav, rho, v0):
    # Put-Call Parity: P = C - S + K*exp(-rT)
    call = H93callvalue(S0, K, T, r, kappav, thetav, sigmav, rho, v0)
    put = call - S0 + K * np.exp(-r * T)
    return max(0.0, float(put))

## Heston and Calibration loop

def H93errorfunction(p):
    kappav, thetav, sigmav, rho, v0 = p
    # Constraint handling for valid Heston parameters
    if kappav <= 0 or thetav <= 0.005 or sigmav <= 0 or rho < -1 or rho > 1 or v0 <= 0:
        return 5000.0
    if (2 * kappav * thetav) < (sigmav**2): # Feller condition
        return 5000.0

    se = []
    # Calculate error for both calls and puts in your 60-day set
    for _, option in df60.iterrows():
        if option["Type"] == "C":
            model_val = H93callvalue(S0, option["Strike"], option["T"], r, kappav, thetav, sigmav, rho, v0)
        else:
            model_val = H93putvalueparity(S0, option["Strike"], option["T"], r, kappav, thetav, sigmav, rho, v0)
        se.append((model_val - option["Price"])**2)

    return np.sqrt(np.sum(se) / len(df60))

# Calibration execution
print("Running Heston calibration...")
# Brute force search to find a good starting region
ranges = (slice(0.1, 5.0, 1.0), slice(0.01, 0.1, 0.02), slice(0.1, 0.5, 0.1), slice(-0.8, 0.0, 0.2), slice(0.01, 0.05, 0.01))
p0 = brute(H93errorfunction, ranges, finish=None)

# Local optimization for precision
opt_heston = fmin(H93errorfunction, p0, xtol=0.0001, ftol=0.0001)


print("Heston defined")
print("Optimal Heston parameters (kappa, theta, sigma_v, rho, v0):")
print(opt_heston)




```

 ## Bates Jump and Characteristics function

```python
def M76Jcharfunc(u, T, lamb, mu_j, sigma_j):
    omega = -lamb * (np.exp(mu_j + 0.5 * sigma_j**2) - 1)
    charfuncvalue = np.exp(1j * u * omega * T + lamb * T * (np.exp(1j * u * mu_j - 0.5 * u**2 * sigma_j**2) - 1))
    return charfuncvalue

def B96charfunc(u, T, r, kappav, thetav, sigmav, rho, v0, lamb, mu_j, sigma_j):

    # Bates = Heston * Jump

    heston = H93charfunc(u, T, r, kappav, thetav, sigmav, rho, v0)
    jump = M76Jcharfunc(u, T, lamb, mu_j, sigma_j)
    return heston * jump

# Defining Bates pricing call

def B96callvalue(S0, K, T, r, kappav, thetav, sigmav, rho, v0, lamb, mu_j, sigma_j):

    # Lewis integral using Bates CF

    def B96intfunc(u):
        charfunc = B96charfunc(u - 1j * 0.5, T, r, kappav, thetav, sigmav, rho, v0, lamb, mu_j, sigma_j)
        return (1 / (u**2 + 0.25)) * np.exp(1j * u * np.log(S0 / K)) * charfunc.real

    intvalue, _ = quad(B96intfunc, 0, np.inf, limit=250)
    callvalue = max(0.0, S0 - np.exp(-r * T) * np.sqrt(S0 * K) / np.pi * intvalue)
    return callvalue

print("Bates")
```

## Defining Error function for Bates calibration


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

df = pd.read_csv("/content/GWP1_Option_data.csv")
df.columns = ["Days", "Strike", "Price", "Type"]
df["T"] = df["Days"] / 250

## Market Inputs
S0 = 232.90
r = 0.015

df60 = df[df["Days"] == 60].copy().reset_index(drop=True)

## Call & Puts
calls60 = df60[df60["Type"] == "C"].sort_values("Strike").reset_index(drop=True)
puts60 = df60[df60["Type"] == "P"].sort_values("Strike").reset_index(drop=True)


def B96errorfunction(p):
    kv, tv, sv, rv, v0v, lb, m, d = p

    # 1. Simple Constraint: Return a high error if parameters go out of bounds
    if kv <= 0 or tv <= 0.005 or sv <= 0 or v0v <= 0 or lb < 0 or d <= 0:
        return 5000.0

    sum_squared_diff = 0
    # 2. Loop through all 60-day options
    for i in range(len(df60)):
        strike = df60.iloc[i]["Strike"]
        maturity = df60.iloc[i]["T"]
        market_price = df60.iloc[i]["Price"]

        # Calculate Bates model price
        model_price = B96callvalue(S0, strike, maturity, r, kv, tv, sv, rv, v0v, lb, m, d)

        # Parity for Puts
        if df60.iloc[i]["Type"] == "P":
            model_price = model_price - S0 + strike * np.exp(-r * maturity)

        sum_squared_diff += (model_price - market_price)**2

    return np.sqrt(sum_squared_diff / len(df60))

# 3. Calibration
# p0 uses your previously calibrated Heston params
p0 = np.concatenate([opt_heston, [0.1, -0.5, 0.1]])

print("Starting Calibration...")
# maxiter limits the run so it doesn't stay stuck
opt_final = fmin(B96errorfunction, p0, xtol=0.001, ftol=0.001, maxiter=50)

print("Calibration Complete!")
print("Optimal Parameters:", opt_final)

# 3. Simple Visual Guide

# Calculate all model prices one last time to compare

model_prices = [B96callvalue(S0, df60.loc[i, "Strike"], df60.loc[i, "T"], r, *opt_final) for i in range(len(df60))]

plt.figure(figsize=(8, 5))
plt.scatter(df60["Strike"], df60["Price"], label="Market Prices (Actual)", color='blue')
plt.plot(df60["Strike"], model_prices, label="Bates Model (Fit)", color='red', linestyle='--')
plt.title("Bates Model fit to market?")
plt.xlabel("Strike Price")
plt.ylabel("Option Price")
plt.legend()
plt.show()
```
